# Reading the CSV & remove Duplicate Rows

```python
import pandas as pd
import numpy as np
file_name = "Clean_reviews.csv"
file_name_output = "Final_Reviews_Clean_withou_dup.csv"
engine='python'
dup_df = pd.read_csv(file_name, sep="\t or ,") 
print("Shape of data Before Removing Duplicates: ",dup_df.shape)
dup_df.drop_duplicates(subset=None, inplace=True)

dup_df.to_csv(file_name_output)
non_dup_df=pd.read_csv("Final_Reviews_Clean_withou_dup.csv")
non_dup_df=non_dup_df.drop('Unnamed: 0',axis=1)
print("Shape of data After Removing Duplicates: ",non_dup_df.shape)
print("Any Null Value: ",non_dup_df.isnull().sum())
non_dup_df.head()
```

# Data Cleaning

```python
features=non_dup_df.iloc[:,0].values
features
```

```python
import re
processed_features = []
#processed_features = [word.lower() for word in features] 
for sentence in range(0, len(features)):
    # Remove all the special characters
    processed_feature = re.sub(r'\W', ' ', str(features[sentence]))

    # remove all single characters
    processed_feature= re.sub(r'\s+[a-zA-Z]\s+', '', processed_feature)
    #Search for all non-letters Replace all non-letters with spaces
    processed_feature = re.sub("[^a-zA-Z]"," ", str(processed_feature))

    # Search for only alphabet
    processed_feature = re.sub(r'\^[a-zA-Z]\s+', '', processed_feature) 

    # Substituting multiple spaces with single space
    processed_feature = re.sub(r'\s+', ' ', processed_feature, flags=re.I)

    # Removing prefixed 'b'
    processed_feature = re.sub(r'^b\s+', '', processed_feature)

    shortword = re.compile(r'\W*\b\w{1,2}\b')
    processed_feature=shortword.sub('', processed_feature)
    # Converting to Lowercase
    processed_feature= processed_feature.lower()
    processed_features.append(processed_feature)

clean_data_file = pd.DataFrame(data=processed_features ,columns=['Reviews'])
clean_data_file.to_csv('data_clean_final.csv', index=None, header=True)
print("Any Null Value: ",clean_data_file.isnull().sum())
print(clean_data_file.shape)
clean_data_file.head()
```

# Remove Non-English Words

```python
import nltk
import nltk.data
from nltk.tokenize import word_tokenize
reviews = clean_data_file.Reviews.str.cat(sep=' ')
#function to split text into word
tokens = word_tokenize(reviews)
vocabulary = set(tokens)
print(len(vocabulary))
frequency_dist = nltk.FreqDist(tokens)
sorted(frequency_dist,key=frequency_dist.__getitem__, reverse=True)
print(vocabulary)
```

# Removing Stop-words 

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if not w in stop_words]
print(tokens)
```

```python
#NLTK provides several stemmer interfaces like Porter stemmer, #Lancaster Stemmer, Snowball Stemmer
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stems = []
for t in tokens:    
    stems.append(porter.stem(t))
print(stems)
```

# Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud().generate_from_frequencies(frequency_dist)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
